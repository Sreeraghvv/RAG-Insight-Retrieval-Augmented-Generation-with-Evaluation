```python
import ollama
import re

EVAL_MODEL = "phi3"

def extract_score(text):
    match = re.search(r"(\d)", text)
    if match:
        return int(match.group(1))
    return None

def evaluate_faithfulness(query, context, answer):
    prompt = f"""
Check if the answer is fully supported by the context.

Context:
{context}

Question:
{query}

Answer:
{answer}

Score from 1 to 5 and explain.
"""

    response = ollama.chat(
        model=EVAL_MODEL,
        messages=[{"role": "user", "content": prompt}],
        options={"temperature": 0, "num_gpu": 0}
    )

    output = response["message"]["content"]
    return extract_score(output), output

def evaluate_relevance(query, answer):
    prompt = f"""
Evaluate how well the answer addresses the question.

Question:
{query}

Answer:
{answer}

Score from 1 to 5 and explain.
"""

    response = ollama.chat(
        model=EVAL_MODEL,
        messages=[{"role": "user", "content": prompt}],
        options={"temperature": 0, "num_gpu": 0}
    )

    output = response["message"]["content"]
    return extract_score(output), output
```
